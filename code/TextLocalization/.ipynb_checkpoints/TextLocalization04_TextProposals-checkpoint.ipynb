### A selective search algorithm for end-to-end word recognition


### 1. Over-segmenation of the input image with the MSER algorithm

```python
# set up Python environment: numpy for numerical routines, and matplotlib for plotting
import numpy as np
import matplotlib.pyplot as plt
# display plots in this notebook
%matplotlib inline

# set display defaults
plt.rcParams['figure.figsize'] = (10, 10)        # large images
plt.rcParams['image.interpolation'] = 'nearest'  # don't interpolate
plt.rcParams['image.cmap'] = 'gray'  # use grayscale output

# load OpenCV python module
import cv2
```

```python
# load the input image
img = cv2.imread('./images/scenetext01_resized24.jpg', 0);

# load a color version for visualization
vis = cv2.imread('./images/scenetext01_resized24.jpg');

# detect Maximally Stable Estremal Regions
mser = cv2.MSER_create()
regions = mser.detectRegions(img)

#Visualization of MSER bounding boxes
rects = [cv2.boundingRect(p) for p in regions[0]]

for rect in rects:
    cv2.rectangle(vis, rect[0:2], (rect[0]+rect[2],rect[1]+rect[3]), (0, 0, 255), 2)

vis = vis[:,:,::-1] #flip the colors dimension from BGR to RGB
plt.imshow(vis)
```

### 2. Generate text groups (e.g. words) candidates with bottom-up agglomerative clustering

```python
# we represent each MSER region in a 3-d feature space by it's bounding box height and center coordinates
region_centers = np.array([[r[0]+r[2]/2, r[1]+r[3]/2] for r in rects]).astype(float)
region_heights = np.array([r[3] for r in rects]).astype(float)[:,None]

region_centers[:,0] = region_centers[:,0]/img.shape[1]
region_centers[:,1] = region_centers[:,1]/img.shape[0]
region_heights[:,0] = region_heights[:,0]/img.shape[0]

X = np.hstack((region_centers,region_heights))


print np.shape(X)
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage

# generate the linkage matrix
Z = linkage(X, 'single')

dendrogram(Z, no_labels=True, color_threshold=15)
plt.show()
```

```python
#Visualization of MSER clusters bounding boxes (nodes in the dendrogram)
for c in Z:
    group_x1 = min(rects[int(c[0])][0],rects[int(c[1])][0])
    group_y1 = min(rects[int(c[0])][1],rects[int(c[1])][1])
    group_x2 = max(rects[int(c[0])][0]+rects[int(c[0])][2],rects[int(c[1])][0]+rects[int(c[1])][2])
    group_y2 = max(rects[int(c[0])][1]+rects[int(c[0])][3],rects[int(c[1])][1]+rects[int(c[1])][3])
    rects.append((group_x1,group_y1,group_x2-group_x1,group_y1-group_y2))

# load a clean image for visualization
vis = cv2.imread('./images/scenetext01_resized24.jpg');

num_regions = np.shape(X)[0]
for r in range(num_regions,np.shape(rects)[0]):
    rect = rects[r]
    cv2.rectangle(vis, rect[0:2], (rect[0]+rect[2],rect[1]+rect[3]), (255, 0, 0), 2)

vis = vis[:,:,::-1] #flip the colors dimension from BGR to RGB
plt.imshow(vis)
```

### 3. Diversification strategies

We have seen how to generate text grouping candidates with a simple hierarchical clustering analysis of MSER regions by their bbox height similarity and their spatial distance. 

We can do the same for a number of other similarity measures: color, stroke width, etc... 

This way we maximize the chances of finding text groups in many different conditions.

Similarly we can increase the overall word detection recall by doing the same agglomerations in different color channels and different scales.
